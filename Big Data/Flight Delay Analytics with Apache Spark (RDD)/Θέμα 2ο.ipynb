```python
spark = SparkSession.builder.master("local[*]").appName("Spark DataFrame").getOrCreate()
print("Spark session started")
```

**Βήμα 1ο**

```python
from google.colab import files
#from google.colab import drive
#drive.mount('/content/drive')

#filename = "/content/drive/My Drive/Colab Notebooks/Big Data/HW1/flights_2000.csv"
uploaded = files.upload()            # pick your CSV
filename = next(iter(uploaded.keys()))
df = spark.read.csv(filename, header=True, inferSchema=True)
print("CSV file loaded into DataFrame")
```

**Βήμα 2ο**

```python
df.printSchema()
```

**Βήμα 3ο**

```python
filtered_df = df.filter(df["CANCELLED"] == 0).na.drop(subset=["DEP_DELAY"])
routes_df = filtered_df.groupBy("ORIGIN_AIRPORT", "DEST_AIRPORT").agg(F.avg("DEP_DELAY").alias("avg_dep_delay"))
top_10_routes = routes_df.orderBy("avg_dep_delay", ascending=False).limit(10)
```

**Βήμα 4ο**

```python
top_10_routes.write.csv("top_10_routes", header=True, mode="overwrite")
```
