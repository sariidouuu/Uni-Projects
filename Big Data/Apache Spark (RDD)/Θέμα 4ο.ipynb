```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded = files.upload()
filename = next(iter(uploaded.keys()))
print("CSV file loaded into Pandas DataFrame")
```

**Βήμα 1ο**

```python
pandas_df = pd.read_csv(filename)
print(pandas_df)
```

**Βήμα 2ο και 3ο**

```python

# --- Προετοιμασία Δεδομένων για το Γράφημα ---
# Συνδυάζουμε τα πεδία ORIGIN_AIRPORT και DEST_AIRPORT για να φτιάξουμε το 'route'
pandas_df['Route'] = pandas_df['ORIGIN_AIRPORT'] + ' -> ' + pandas_df['DEST_AIRPORT']

# Δεδομένα για τους άξονες
routes = pandas_df['Route']
delays = pandas_df['avg_dep_delay']

# Γράφημα με matplotlib
plt.figure(figsize=(10, 7))

# --- 3. Εμφανίστε τα 10 routes με τη μεγαλύτερη μέση καθυστέρηση (οριζόντιο γράφημα). ---
plt.barh(routes, delays, color='lightcoral') # Άλλαξα το χρώμα για να ξεχωρίζει

# Προσθήκη ετικετών και τίτλου
plt.xlabel('Μέση Καθυστέρηση Αναχώρησης (λεπτά)')
plt.ylabel('Διαδρομή (Route)')
plt.title('Top 10 Διαδρομές με τη Μεγαλύτερη Μέση Καθυστέρηση')

# Αντιστροφή του άξονα Y για να είναι η μεγαλύτερη καθυστέρηση πάνω-πάνω
plt.gca().invert_yaxis()

# Εμφάνιση πλέγματος (προαιρετικό, για ομορφιά)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Εμφάνιση του γραφήματος
plt.tight_layout()
plt.show()
```
